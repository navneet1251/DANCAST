```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Reshape
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from skimage.color import rgb2gray
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from PIL import Image

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

class Config:
    IMG_SIZE = (300, 300)
    GRAYSCALE = True
    BATCH_SIZE = 32
    EPOCHS = 50
    EARLY_STOPPING_PATIENCE = 3
    MIN_DELTA = 0.001
    INIT_FILTERS = 48
    VALIDATION_SPLIT = 0.2
    TRAIN_SAMPLES = 2875
    TEST_SAMPLES = 715

class GlobalModels:
    kmeans = None

def build_autoencoder():
    # Encoder
    input_img = Input(shape=(*Config.IMG_SIZE, 1))

    # Layer 1: Conv (48 filters)
    x = Conv2D(Config.INIT_FILTERS, (3, 3), activation='relu', padding='same')(input_img)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(Config.INIT_FILTERS//2, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(Config.INIT_FILTERS//4, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    x = Conv2D(Config.INIT_FILTERS//8, (3, 3), activation='relu', padding='same')(x)
    encoded = MaxPooling2D((2, 2), padding='same')(x)

    # Decoder
    x = Flatten()(encoded)
    x = Dense(256, activation='relu')(x)
    x = Dense(512, activation='relu')(x)
    decoded = Dense(Config.IMG_SIZE[0] * Config.IMG_SIZE[1], activation='sigmoid')(x)
    decoded = Reshape((*Config.IMG_SIZE, 1))(decoded)

    autoencoder = Model(input_img, decoded)
    autoencoder.compile(optimizer=Adam(), loss='mse')
    return autoencoder

class CastingDataLoader:
    def __init__(self, data_dir):
        self.data_dir = data_dir
        self.image_paths = self._load_image_paths()

    def _load_image_paths(self):
        paths = []
        for root, _, files in os.walk(self.data_dir):
            for file in files:
                if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                    paths.append(os.path.join(root, file))
        if not paths:
            raise ValueError(f"No images found in {self.data_dir}")
        return paths

    def load_images(self, max_samples=None):
        images = []
        for i, path in enumerate(self.image_paths[:max_samples] if max_samples else self.image_paths):
            try:
                # Load and resize image
                img = Image.open(path).resize(Config.IMG_SIZE)
                img = np.array(img) / 255.0

                # Convert to grayscale
                if Config.GRAYSCALE:
                    img = rgb2gray(img).reshape(*Config.IMG_SIZE, 1)

                images.append(img)
            except Exception as e:
                print(f"Error loading {path}: {e}")
        return np.array(images)

def train_autoencoder(non_defective_dir):
    # Train on non-defective images
    loader = CastingDataLoader(non_defective_dir)
    images = loader.load_images(max_samples=Config.TRAIN_SAMPLES)

    if len(images) == 0:
        raise ValueError("No training images found. Check data directory.")

    # Build autoencoder
    autoencoder = build_autoencoder()
    autoencoder.summary()

    # Early stopping (δ=0.001, patience=3)
    early_stopping = EarlyStopping(
        monitor='loss',
        min_delta=Config.MIN_DELTA,
        patience=Config.EARLY_STOPPING_PATIENCE,
        restore_best_weights=True
    )

    history = autoencoder.fit(
        images, images,
        batch_size=Config.BATCH_SIZE,
        epochs=Config.EPOCHS,
        shuffle=True,
        callbacks=[early_stopping],
        validation_split=Config.VALIDATION_SPLIT
    )

    # Plot training history
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.ylabel('MSE Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.show()

    return autoencoder

def calculate_mae(autoencoder, images):
    mae_scores = []
    for img in images:
        reconstructed = autoencoder.predict(np.expand_dims(img, axis=0))[0]
        mae = np.mean(np.abs(img - reconstructed))
        mae_scores.append(mae)
    return np.array(mae_scores)

def evaluate_model(autoencoder, test_defective_dir, test_non_defective_dir):

    def_loader = CastingDataLoader(test_defective_dir)
    non_def_loader = CastingDataLoader(test_non_defective_dir)
    def_images = def_loader.load_images()
    non_def_images = non_def_loader.load_images()

    # Calculate MAE
    def_mae = calculate_mae(autoencoder, def_images)
    non_def_mae = calculate_mae(autoencoder, non_def_images)

    # Combine scores and create labels (1=defective, 0=non-defective)
    all_mae = np.concatenate([def_mae, non_def_mae])
    true_labels = np.concatenate([np.ones(len(def_mae)), np.zeros(len(non_def_mae))])

    kmeans = KMeans(n_clusters=2, random_state=42)
    cluster_labels = kmeans.fit_predict(all_mae.reshape(-1, 1))
    GlobalModels.kmeans = kmeans

    adjusted_labels = adjust_clusters(all_mae, cluster_labels)

    # Evaluate performance
    print("\nClassification Report:")
    print(classification_report(true_labels, adjusted_labels,
                              target_names=["Non-Defective", "Defective"]))

    cm = confusion_matrix(true_labels, adjusted_labels)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                display_labels=["Non-Defective", "Defective"])
    disp.plot(cmap='Blues')
    plt.title("Confusion Matrix")
    plt.show()

    return all_mae, true_labels, adjusted_labels

def adjust_clusters(mae_scores, cluster_labels):

    cluster0_avg = np.mean(mae_scores[cluster_labels == 0])
    cluster1_avg = np.mean(mae_scores[cluster_labels == 1])


    if cluster0_avg > cluster1_avg:
        return np.where(cluster_labels == 0, 1, 0)
    return cluster_labels

def visualize_results(autoencoder, images, labels, mae_scores, num_samples=5):

    def_indices = np.where(labels == 1)[0][:num_samples]
    non_def_indices = np.where(labels == 0)[0][:num_samples]

    plt.figure(figsize=(15, 10))
    for i, idx in enumerate(def_indices):
        img = images[idx]
        reconstructed = autoencoder.predict(np.expand_dims(img, axis=0))[0]

        # Original image
        plt.subplot(4, num_samples, i+1)
        plt.imshow(img.squeeze(), cmap='gray')
        plt.title(f"Defective (MAE: {mae_scores[idx]:.4f})")
        plt.axis('off')

        # Reconstructed image
        plt.subplot(4, num_samples, i+num_samples+1)
        plt.imshow(reconstructed.squeeze(), cmap='gray')
        plt.title("Reconstructed")
        plt.axis('off')

    for i, idx in enumerate(non_def_indices):
        img = images[idx]
        reconstructed = autoencoder.predict(np.expand_dims(img, axis=0))[0]

        # Original image
        plt.subplot(4, num_samples, i+2*num_samples+1)
        plt.imshow(img.squeeze(), cmap='gray')
        plt.title(f"Non-Defective (MAE: {mae_scores[idx]:.4f})")
        plt.axis('off')

        # Reconstructed image
        plt.subplot(4, num_samples, i+3*num_samples+1)
        plt.imshow(reconstructed.squeeze(), cmap='gray')
        plt.title("Reconstructed")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

def predict_defect(autoencoder, image_path):
    if GlobalModels.kmeans is None:
        raise ValueError("KMeans model not trained. Run evaluation first.")

    try:
        img = Image.open(image_path).resize(Config.IMG_SIZE)
        img = np.array(img) / 255.0

        if Config.GRAYSCALE:
            img = rgb2gray(img).reshape(*Config.IMG_SIZE, 1)

        reconstructed = autoencoder.predict(np.expand_dims(img, axis=0))[0]
        mae_score = np.mean(np.abs(img - reconstructed))

        cluster = GlobalModels.kmeans.predict([[mae_score]])[0]

        cluster0_avg = GlobalModels.kmeans.cluster_centers_[0][0]
        cluster1_avg = GlobalModels.kmeans.cluster_centers_[1][0]

        if cluster0_avg > cluster1_avg:
            prediction = 1 if cluster == 0 else 0
        else:
            prediction = 1 if cluster == 1 else 0

        return prediction, mae_score

    except Exception as e:
        print(f"Error processing image: {e}")
        return None, None

if __name__ == "__main__":
    train_non_defective_dir = "/content/drive/MyDrive/casting_data/train/ok_front"
    test_defective_dir = "/content/drive/MyDrive/casting_data/test/def_front"
    test_non_defective_dir = "/content/drive/MyDrive/casting_data/test/ok_front"

    try:
        print("Training autoencoder on non-defective images...")
        autoencoder = train_autoencoder(train_non_defective_dir)

        print("\nEvaluating on test set...")
        mae_scores, true_labels, pred_labels = evaluate_model(
            autoencoder,
            test_defective_dir,
            test_non_defective_dir
        )

        print("\nVisualizing results...")
        def_loader = CastingDataLoader(test_defective_dir)
        non_def_loader = CastingDataLoader(test_non_defective_dir)
        test_images = np.concatenate([def_loader.load_images(), non_def_loader.load_images()])
        visualize_results(autoencoder, test_images, true_labels, mae_scores)

        print("\nTesting individual image prediction...")
        test_image_path = input("Enter path to image for defect prediction: ").strip()
        if os.path.exists(test_image_path):
            prediction, mae = predict_defect(autoencoder, test_image_path)
            if prediction is not None:
                print(f"\nPrediction for {test_image_path}:")
                print(f"MAE Score: {mae:.4f}")
                print("Classification:", "DEFECTIVE" if prediction == 1 else "NON-DEFECTIVE")
            else:
                print("Failed to make prediction for the image.")
        else:
            print(f"File not found: {test_image_path}")

    except Exception as e:
        print(f"Error: {e}")
        print("\nTroubleshooting tips:")
        print("1. Verify dataset paths are correct")
        print("2. Check images exist in specified directories")
        print(f"Current paths:")
        print(f"- Training: {train_non_defective_dir}")
        print(f"- Test Defective: {test_defective_dir}")
        print(f"- Test Non-Defective: {test_non_defective_dir}")
```
